```python
a = np.linspace(0,10,11)
```

```python
b = a**4
```

```python
a
```

```python
x = np.arange(0,11)
```

```python
y = 2*x
```

```python
fig = plt.figure()
axes = fig.add_axes([0.2,0.2,0.3,0.3])
axes.plot(a,b)
axes.set_xlim(0,8)
axes.set_ylim(0,8000)
axes.set_xlabel('Aroy')
axes.set_ylabel('Bory')
axes.set_title('PowerFire')
yxes = fig.add_axes([0.5,0.5,1,1])
yxes.plot(x,y)
yxes.set_xlim(0,10)
yxes.set_ylim(0,20)
yxes.set_xlabel('Xtin')
yxes.set_ylabel('Ytin')
yxes.set_title('Stagerun')
fig.suptitle()
fig.savefig('hoshera.png',bbox_inches='tight')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=1)
axes.plot(a,b)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(a,b)
fig.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][0].set_xlabel('First')
axes[0][0].set_ylabel('second')
axes[0][0].set_xlim(0,14)
axes[0][0].set_ylim(0,18)
axes[1][0].plot(a,b)
axes[1][0].set_xlabel('neto')
axes[1][0].set_ylabel('kory')
axes[1][0].set_xlim(0,18)
axes[1][0].set_ylim(1,800)
axes[0][1].plot(x,y)
axes[0][1].set_xlabel('otyr')
axes[0][1].set_ylabel('neetg')
axes[1][1].plot(a,b)
axes[1][1].set_xlabel('hoid')
axes[1][1].set_ylabel('fjoe')
fig.suptitle('Ninth Form')
fig.tight_layout()
fig.subplots_adjust(wspace=0.7,hspace=1)
fig.savefig('New_subplots',bbox_inches='tight')
```

```python
import numpy as np
import matplotlib.pyplot as plt
x= np.arange(10)
y= 2*x
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,label='UND')
axes.plot(x,y*2,label='BNR')
axes.legend()
```
